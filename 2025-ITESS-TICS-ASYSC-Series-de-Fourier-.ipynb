## DESCRIPCION APROXIMACION DE UNA ONDA CUADRADA APROXIMACION DE SEÑALES PERIODICAS USANDO SERIES DE FOURIER
$$ s(t) = a_0 + \sum_{n=1}^{\infty} [ a_n cos(2 \pi f n t ) + b_n sin(2 \pi f n t ) ]  $$

$$
s(f) = a_0 + \sum_{n=1}^{\infty} \left[ a_n \cos(2\pi f_n t) + b_n \sin(2\pi f_n t) \right]
$$

## donde los coeficientes están dados por:

$$
a_0 = \frac{1}{T} \int_{0}^{T} f(t)\, dt
$$

$$
a_n = \frac{2}{T} \int_{0}^{T} f(t)\cos(2\pi f_n t)\, dt
$$

$$
b_n = \frac{2}{T} \int_{0}^{T} f(t)\sin(2\pi f_n t)\, dt
$$

```python
from thinkdsp import CosSignal, SinSignal

f = 0.25
k = 1
st = CosSignal(freq=f, amp=2*k/3.1416, offset=0)
```

```python
from thinkdsp import decorate
wave = st.make_wave(duration=4)
wave.ys += 0.5

wave.plot()
decorate(xlabel='Time (s)')
```
